```python
from google.colab import files
import pandas as pd, re

uploaded = files.upload()
fname = next(iter(uploaded))

pattern = re.compile(
    r'(?P<ip>\S+)\s+\S+\s+\S+\s'
    r'\[(?P<datetime>.*?)\]\s'
    r'"(?P<method>\S+)\s(?P<endpoint>\S+)\s(?P<protocol>[^"]+)"\s'
    r'(?P<status>\d{3})\s(?P<size>\S+)\s'
    r'"(?P<referrer>[^"]*)"\s"(?P<user_agent>[^"]*)"'
)

rows = []
with open(fname, 'r', encoding='utf-8', errors='ignore') as f:
    for line in f:
        m = pattern.search(line)
        if m:
            rows.append(m.groupdict())

df = pd.DataFrame(rows)

df['size'] = pd.to_numeric(df['size'].replace('-', '0'), errors='coerce')
df['status'] = pd.to_numeric(df['status'], errors='coerce')
df['datetime'] = pd.to_datetime(df['datetime'], format='%d/%b/%Y:%H:%M:%S %z', errors='coerce')

print(df.head())
```

```python
df.to_csv('access_clean.csv', index=False)
```
