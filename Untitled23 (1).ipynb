```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
df = pd.read_excel(list(uploaded.keys())[0])
df.head()
```

```python
df = df.dropna(subset=['CustomerID'])
df = df[df['Quantity'] > 0]
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df.head()
```

```python
snapshot = df['InvoiceDate'].max() + pd.Timedelta(days=1)

rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot - x.max()).days,
    'InvoiceNo': 'nunique',
    'TotalPrice': 'sum'
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']
rfm.head()


```

```python
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm)
rfm_scaled = pd.DataFrame(rfm_scaled, columns=rfm.columns, index=rfm.index)
rfm_scaled.head()


```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

wcss = []   # Within-Cluster Sum of Square
silhouette_scores = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(rfm_scaled)
    wcss.append(kmeans.inertia_)

    if k > 1:
        silhouette_scores.append(silhouette_score(rfm_scaled, kmeans.labels_))

# Graphiques
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Mthode du coude
ax1.plot(k_range, wcss, 'bo-', linewidth=2, markersize=8)
ax1.set_xlabel('Nombre de clusters (k)')
ax1.set_ylabel('WCSS (Within-Cluster Sum of Square)')
ax1.set_title('Mthode du Coude')
ax1.grid(True, alpha=0.3)

# Score de silhouette
ax2.plot(range(2, 11), silhouette_scores, 'ro-', linewidth=2, markersize=8)
ax2.set_xlabel('Nombre de clusters (k)')
ax2.set_ylabel('Score de Silhouette')
ax2.set_title('Score de Silhouette')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# 3.2 Dtermination du nombre optimal de clusters
from sklearn.metrics import silhouette_score
wcss = []   # Within-Cluster Sum of Square
silhouette_scores = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(rfm_scaled)
    wcss.append(kmeans.inertia_)

    if k > 1:
        silhouette_scores.append(silhouette_score(rfm_scaled, kmeans.labels_))
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)
rfm.head()
```

```python
rfm.groupby('Cluster').mean()
```

```python
plt.scatter(rfm['Recency'], rfm['Monetary'], c=rfm['Cluster'])
plt.xlabel("Recency")
plt.ylabel("Monetary")
plt.title("Clustering des clients")
plt.show()
```
