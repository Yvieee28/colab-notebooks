```python
X = df['Profit']
y = df['R&D Spend']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
import statsmodels.api as sm
```

```python
# Ajout de la constante pour avoir l'intercept
X_train_sm = sm.add_constant(X_train)
```

```python
# Entrainement : calcul des coefficients de la rgression simple
lr = sm.OLS(y_train, X_train_sm).fit()
```

```python
# Affichage des coefficients calculs par le modle
lr.params
```

```python
print(lr.summary())
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, -4.831e+04 +  1.0985*X_train, 'r')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
# Ajout de la constate  X_test
X_test_sm = sm.add_constant(X_test)

# Predire les valeurs de l'ensemble de test
y_pred = lr.predict(X_test_sm)
```

```python
import numpy as np
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
r_squared = r2_score(y_test, y_pred)
r_squared
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, -4.831e+04  + 1.0985  * X_test, 'r')
plt.show()
```

```python
y_train_pred = lr.predict(X_train_sm)
res = (y_train - y_train_pred)
```
